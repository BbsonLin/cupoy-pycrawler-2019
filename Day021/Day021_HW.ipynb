## 1. 取出今天所有的新聞

```python
import time

import datetime
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

today_date = datetime.date.today()

move_page_down = ActionChains(browser).send_keys(Keys.END)

while True:
    time.sleep(2)
    move_page_down.perform()
    html_source = browser.page_source
    soup = BeautifulSoup(html_source)

    last_news = soup.find(class_="part_list_2").find_all('h3')[-1]
    last_news_date = datetime.datetime.strptime(last_news.find(class_="date").text, '%Y/%m/%d %H:%M')
    print(last_news_date.date(), last_news_date.time())
    # 找到時間為昨天的新聞為止
    if last_news_date.date() < today_date:
        break
```

```python
for news in soup.find(class_="part_list_2").find_all('h3'):
    news_datetime = datetime.datetime.strptime(news.find(class_="date").text, '%Y/%m/%d %H:%M')
    if news_datetime.date() < today_date:
        break
    else:
        print(news_datetime, news.find_all('a')[-1].text)
        
browser.quit()
```

## 2. 取出現在時間兩小時內的新聞

```python
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys

browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

two_hours_ago_datetime = datetime.datetime.today() - datetime.timedelta(hours=2)

move_page_down = ActionChains(browser).send_keys(Keys.END)

while True:
    time.sleep(2)
    move_page_down.perform()
    html_source = browser.page_source
    soup = BeautifulSoup(html_source)

    last_news = soup.find(class_="part_list_2").find_all('h3')[-1]
    last_news_date = datetime.datetime.strptime(last_news.find(class_="date").text, '%Y/%m/%d %H:%M')
    print(last_news_date.date(), last_news_date.time())
    # 找到時間為兩個小時前的時間為止
    if last_news_date < two_hours_ago_datetime:
        break
```

```python
for news in soup.find(class_="part_list_2").find_all('h3'):
    news_datetime = datetime.datetime.strptime(news.find(class_="date").text, '%Y/%m/%d %H:%M')
    if news_datetime < two_hours_ago_datetime:
        break
    else:
        print(news_datetime, news.find_all('a')[-1].text)

browser.quit()
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select

browser = webdriver.Chrome(executable_path='./chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")

selector_year = Select(browser.find_element_by_id("selY"))
selector_month = Select(browser.find_element_by_id("selM"))
selector_day = Select(browser.find_element_by_id("selD"))

three_days_ago = datetime.datetime.today() - datetime.timedelta(days=3)
three_days_ago_date = three_days_ago.date()

selector_year.select_by_value(str(three_days_ago_date.year))
selector_month.select_by_value(str(three_days_ago_date.month))
selector_day.select_by_value(str(three_days_ago_date.day))

# 選取日期並查詢
browser.find_element_by_id('button').click()

move_page_down = ActionChains(browser).send_keys(Keys.END)

for _ in range(30):
    time.sleep(2)
    # 向下滾動 30 次，30 次 END Key
    move_page_down.perform()
    
html_source = browser.page_source
# print(three_days_ago_date.year, three_days_ago_date.month, three_days_ago_date.day)
```

```python
soup = BeautifulSoup(html_source)

for news in soup.find(class_="part_list_2").find_all('h3'):
    news_datetime = datetime.datetime.strptime(news.find(class_="date").text, '%Y/%m/%d %H:%M')
    news_time = news_datetime.time()
    # 印出時間為 15 ~ 17 點的新聞
    if news_time.hour >= 15 and news_time.hour <= 17:
        print(news_datetime, news.find_all('a')[-1].text)
    else:
        continue
        
browser.quit()
```

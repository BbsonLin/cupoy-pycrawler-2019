### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text)

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find(class_="tag").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news_dict = {}

for d in soup.find(class_="part_list_2").find_all('h3'):
    tag = d.find(class_="tag").text
    title = d.find_all('a')[-1].text
    if tag not in news_dict.keys():
        news_dict.update({tag: []})
        news_dict[tag].append(title)
    else:
        news_dict[tag].append(title)

print(news_dict)
```

### ③ 哪一個類別的文章最多

```python
length_dict = {key: len(value) for key, value in news_dict.items()}
print(length_dict)
```
